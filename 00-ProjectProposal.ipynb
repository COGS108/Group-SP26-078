## Research Question

Our project asks: Among Spotify tracks across 125 genres, how well do audio features predict a track’s Spotify popularity score, and are those same audio features associated with whether a song appears in the Most Streamed Spotify Songs 2024 dataset? We will use Spotify popularity score as the main outcome variable and audio features such as danceability, energy, valence, tempo, acousticness, speechiness, instrumentalness, liveness, loudness, and duration as predictor variables. We will use regression analysis to predict popularity score, then compare audio features between songs that appear in the 2024 most-streamed dataset and the broader Spotify catalog.



## Background and Prior Work

Spotify is a major music streaming platform, so its data can be useful for studying what kinds of songs become popular. In this project, we focus on Spotify’s popularity score, which is a number from 0 to 100. According to Spotify’s Web API documentation, this score is based mostly on how many plays a track has and how recent those plays are.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Because of this, popularity score is not exactly the same as total streams, but it is still a useful measure for comparing songs on Spotify.

Many previous Spotify-related data projects look at audio features like danceability, energy, valence, tempo, acousticness, and loudness. The Spotify Tracks Dataset on Kaggle includes these features, along with popularity scores and genre labels for a large collection of tracks.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) A Hugging Face version of the same dataset also shows that the data includes about 114,000 tracks across 125 genres, which makes it large enough for exploratory analysis and regression modeling.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) These kinds of datasets are useful because they let us compare measurable song features instead of only looking at artist names or rankings.

One issue with this topic is that song popularity is probably not caused only by audio features. A song can become popular because of the artist, playlist placement, marketing, TikTok trends, or release timing. Spotify’s documentation also notes that popularity depends partly on recent plays, so older songs and currently trending songs may not be judged in exactly the same way.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) This is a limitation of our project because we are not measuring every possible factor that affects popularity.

Our project also uses the Most Streamed Spotify Songs 2024 dataset to compare recent hit songs with the broader Spotify catalog. This dataset includes information such as track name, artist, release date, track score, Spotify streams, and other platform-related metrics for highly streamed songs in 2024.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) We want to use this dataset to see whether songs that became especially streamed in 2024 have different audio features from general Spotify tracks. However, matching songs across two datasets by track name and artist may not be perfect, so we will need to clean the text carefully and check the quality of the match.

Overall, our project builds on previous Spotify data analyses by combining two related questions. First, we will test how well audio features predict Spotify popularity score in the broader Spotify Tracks Dataset. Second, if the datasets can be matched well enough, we will compare whether songs appearing in the 2024 most-streamed dataset have different audio features from general Spotify tracks.

## Hypothesis

We hypothesize that audio features will moderately predict Spotify popularity score, but they will not explain popularity completely. We expect danceability, energy, and loudness to be positively related to popularity. We also expect songs in the Most Streamed Spotify Songs 2024 dataset to have higher average danceability and energy than the broader Spotify catalog.

## Data

Our ideal dataset would have one row for each Spotify track and columns for track name, artist, genre, release date, popularity score, total Spotify streams, and audio features such as danceability, energy, valence, tempo, acousticness, speechiness, instrumentalness, liveness, loudness, and duration. Ideally, the dataset would include at least 100,000 songs across many genres and release years so that the regression analysis is not based on a small or narrow sample. These data would be collected directly from Spotify or Spotify’s API and stored as a tidy CSV file, where each row is a track and each column is one variable.

For our real datasets, we plan to use the Spotify Tracks Dataset from Kaggle as our main dataset. This dataset is located at https://www.kaggle.com/datasets/maharshipandya/-spotify-tracks-dataset and can be downloaded as a CSV file without special permission. It contains Spotify tracks across 125 genres and includes audio features associated with each track, such as danceability, energy, valence, tempo, acousticness, instrumentalness, loudness, and popularity score. This dataset will let us use regression analysis to test how well audio features predict Spotify popularity score. The same dataset is also described on Hugging Face as a CSV dataset that can be used for recommendation, classification, and other analysis tasks. 

We also plan to use the Most Streamed Spotify Songs 2024 dataset as a comparison dataset. This dataset is located at https://www.kaggle.com/datasets/nelgiriyewithana/most-streamed-spotify-songs-2024 and includes highly streamed songs from 2024, with variables such as track name, artist, release date, track score, and Spotify streams. We will use this dataset to identify songs that appear among the most-streamed tracks of 2024. If matching by track name and artist works well enough, we will compare the audio features of these 2024 hit songs with the broader Spotify catalog. A possible limitation is that song names and artist names may not match perfectly across the two datasets because of spelling differences, featured artists, remixes, or different versions of the same song.
