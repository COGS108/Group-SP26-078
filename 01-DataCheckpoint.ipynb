## Research Question

Our project asks: To what extent can Spotify audio features predict or explain a track's Spotify popularity score? We will use Spotify popularity score as the main outcome variable, and we will use audio features such as danceability, energy, valence, tempo, acousticness, speechiness, instrumentalness, liveness, loudness, and duration as the main predictor variables. Through regression analysis, we want to understand whether these audio features have a clear relationship with popularity score and which features may be more strongly associated with popularity.

In addition to the main popularity prediction question, we will also use the Most Streamed Spotify Songs 2024 dataset as a comparison dataset. This dataset helps us identify recent hit songs. We will try to match songs from the 2024 dataset to the Spotify Tracks Dataset using track name and artist. If the matching process works well enough, we will compare the audio features of these 2024 hit songs with songs from the broader Spotify track sample.

In this project, genre is not our main research target, and we are not trying to predict a song's genre. Genre will only be used as a background variable to help us understand the dataset, check possible genre imbalance, and explain limitations in the discussion. Our main question is still about the relationship between Spotify audio features and popularity score, and whether hit songs show different audio-feature patterns from the broader track sample.

## Background and Prior Work

Spotify is one of the major music streaming platforms, so Spotify data can be useful for studying the relationship between song features and popularity. In this project, we focus on Spotify popularity score. Spotify's Web API describes track popularity as a value from 0 to 100, with 100 being the most popular. The score is based mostly on the total number of plays a track has and how recent those plays are. Therefore, popularity score is not exactly the same as total lifetime streams, but it is still a useful outcome variable for comparing tracks on Spotify.

Prior research suggests that audio features can help explain part of song popularity. For example, Saragih's study, "Predicting Song Popularity Based on Spotify's Audio Features," uses Spotify audio features and machine learning methods to study song popularity prediction. This supports our choice to use variables such as danceability, energy, loudness, tempo, valence, and acousticness as predictors. These variables describe parts of a song's rhythm, energy, volume, mood, and sound style, so they may be related to how often listeners play a song.

However, predicting hit songs is not a simple problem. Pachet and Roy's work, "Hit Song Science Is Not Yet a Science," argues that predicting song popularity from acoustic or human-labeled features is difficult. This point is important for our project because a song's popularity is not only determined by the sound of the music itself. Other factors, such as artist fame, playlist placement, marketing, release timing, TikTok trends, and platform recommendation systems, can also affect popularity. Many of these factors are not fully captured by the audio features in the Spotify Tracks Dataset.

Other research also shows that Spotify audio features are useful for describing music. Duman et al. discuss Spotify features such as danceability, energy, valence, tempo, loudness, speechiness, and acousticness as measurable ways to describe musical properties and listening experiences. This supports our use of Spotify audio features as meaningful variables, but it does not mean these features directly cause popularity.

The research gap our project addresses is how much Spotify audio features can actually explain popularity score. These features may provide useful prediction information, but they may only explain part of why a song becomes popular. We also want to know whether 2024 hit songs have different audio-feature patterns from the broader Spotify track sample. Our project will not claim that audio features cause popularity. Instead, we will study the association between audio features and popularity score, and we will discuss the limits of this relationship.

## Hypothesis


We hypothesize that Spotify audio features are associated with Spotify popularity score, but this relationship will be moderate rather than strong. Specifically, we expect danceability, energy, and loudness to show positive relationships with popularity score. In other words, songs that are more danceable, energetic, and louder may have higher popularity scores on average.

This hypothesis is based on the idea that these audio features may make songs more suitable for repeated listening, playlists, social settings, and mainstream streaming environments. Danceable songs may be easier to play in social or casual listening contexts. Energetic songs may feel more engaging to listeners. Louder songs may also stand out more in streaming and playlist settings. For these reasons, we expect these features to have some connection with popularity.

As a supplementary comparison, we also expect matched songs from the Most Streamed Spotify Songs 2024 dataset to have higher average danceability, energy, and loudness than the broader Spotify track sample. However, this comparison depends on whether enough songs can be matched successfully between the two datasets.

We do not expect audio features to fully explain popularity. Popularity is also affected by non-audio factors such as artist fame, playlist placement, promotion, TikTok trends, release timing, and platform recommendations. Since these factors are not fully captured in our audio-feature dataset, our expectation is that audio features will explain only part of the variation in popularity score.

### Dataset #1: Spotify Tracks Dataset

This dataset contains 114,000 tracks collected from Spotify-related track data and organized across 114 music genres with 1,000 tracks per genre. Each row represents one track-genre pairing, meaning the same track can appear more than once if it is tagged under multiple genres. After dropping the row with missing track information and removing duplicate `track_id` entries, the cleaned dataset has 89,740 unique tracks.

**Key variables and their meanings:**

- `popularity`: Integer from 0 to 100. Spotify's internal popularity score, calculated based mostly on total play count and how recent those plays are. A score of 100 is the most popular. This is our main outcome variable.
- `danceability`: Float from 0.0 to 1.0. Describes how suitable a track is for dancing, based on tempo, rhythm stability, beat strength, and regularity. Higher values indicate more danceable tracks.
- `energy`: Float from 0.0 to 1.0. Measures perceived intensity and activity. High-energy tracks feel fast, loud, and noisy; low-energy tracks feel calm and quiet.
- `loudness`: Overall loudness of the track in decibels (dB). Less negative values usually mean louder tracks.
- `valence`: Float from 0.0 to 1.0. Describes musical positiveness. High valence means the track sounds happier or more cheerful, while low valence means it sounds sadder, angrier, or more tense.
- `tempo`: Estimated beats per minute (BPM). Most normal songs fall roughly between 50 and 220 BPM, but we will check unusual values during cleaning.
- `acousticness`: Float from 0.0 to 1.0. Confidence that the track is acoustic. Values close to 1.0 mean the track is likely acoustic.
- `speechiness`: Float from 0.0 to 1.0. Detects spoken words. Values above 0.66 suggest mostly spoken word content; values below 0.33 suggest music with little or no speech.
- `instrumentalness`: Float from 0.0 to 1.0. Predicts whether the track has no vocals. Values above 0.5 suggest the track is likely instrumental.
- `liveness`: Float from 0.0 to 1.0. Detects whether a live audience is present. Values above 0.8 strongly suggest a live recording.
- `duration_ms`: Track duration in milliseconds. Most tracks fall between 60,000 ms (1 minute) and 600,000 ms (10 minutes), but very short or very long tracks will be checked during cleaning.
- `explicit`: Boolean. True if the track contains explicit lyrics.
- `key`: Integer from 0 to 11. Estimated overall key of the track (0 = C, 1 = C#/Db, …, 11 = B).
- `mode`: 1 = major, 0 = minor.
- `time_signature`: Estimated number of beats per measure. In this dataset, common values include 3, 4, and 5, while unusual values such as 0 or 1 need to be checked.
- `track_genre`: Categorical. One of 114 genre labels assigned to the track.

**Data concerns:** The genre-balanced design, with 1,000 tracks per genre, means results may not generalize to the actual genre distribution on Spotify. Popularity scores are also skewed toward lower values for many tracks, suggesting the dataset includes many low-visibility songs. One row has missing values in `artists`, `album_name`, and `track_name`, and it is dropped during cleaning. Another important concern is duplication: many tracks appear under multiple genres with the same `track_id`, so we remove duplicate `track_id` entries before the main analysis.

```python
import pandas as pd
import numpy as np
import os

# --- Load raw data ---
tracks_raw = pd.read_csv('data/00-raw/spotify_tracks.csv')
print('Raw shape:', tracks_raw.shape)
print('Columns:', list(tracks_raw.columns))

# --- Missing values ---
print('\nMissing values per column:')
print(tracks_raw.isnull().sum()[tracks_raw.isnull().sum() > 0])

# --- Check duplicate track_ids ---
print(f'\nDuplicate track_id rows: {tracks_raw["track_id"].duplicated().sum()}')
print('(Same track can appear under multiple genres, so duplicate track_id rows need to be removed.)')

# --- Clean ---
tracks = tracks_raw.drop(columns=['Unnamed: 0'])

# Drop the row with missing artist/album/track name
tracks = tracks.dropna(subset=['artists', 'album_name', 'track_name'])

# Keep only one row per track_id
# The first occurrence preserves one genre label for each unique track
tracks = tracks.drop_duplicates(subset=['track_id']).reset_index(drop=True)

print('\nCleaned shape:', tracks.shape)
print('Missing values after cleaning:', tracks.isnull().sum().sum())

# --- Key variable summaries ---
print('\nPopularity summary:')
print(tracks['popularity'].describe().round(2))

features_01 = [
    'danceability', 'energy', 'valence', 'acousticness',
    'speechiness', 'instrumentalness', 'liveness'
]

features_other = ['loudness', 'tempo', 'duration_ms']

print('\n0-1 audio feature ranges:')
print(tracks[features_01].agg(['min', 'max', 'mean']).round(3))

print('\nOther audio-related feature ranges:')
print(tracks[features_other].agg(['min', 'max', 'mean']).round(3))

# --- Check expected ranges / suspicious values ---
print('\nRange checks:')
print('Popularity outside 0-100:', ((tracks['popularity'] < 0) | (tracks['popularity'] > 100)).sum())

for col in features_01:
    invalid_count = ((tracks[col] < 0) | (tracks[col] > 1)).sum()
    print(f'{col} outside 0-1:', invalid_count)

print('Duration <= 0:', (tracks['duration_ms'] <= 0).sum())
print('Tempo <= 0:', (tracks['tempo'] <= 0).sum())
print('Loudness > 0:', (tracks['loudness'] > 0).sum())

# --- Save cleaned dataset ---
os.makedirs('data/02-processed', exist_ok=True)
tracks.to_csv('data/02-processed/spotify_tracks_clean.csv', index=False)

print('\nSaved: data/02-processed/spotify_tracks_clean.csv')
print('Final shape:', tracks.shape)
```

After cleaning, the Spotify Tracks Dataset has 89,740 rows and 20 columns. We removed the extra index column, dropped the one row with missing track information, and removed duplicate `track_id` rows so each track only appears once in the main dataset.

The main outcome variable, `popularity`, is valid from 0 to 100. Its mean is about 33.20 and the median is 33, so many songs in this dataset have relatively low popularity scores. The 0–1 audio features, such as `danceability`, `energy`, `valence`, `acousticness`, `speechiness`, `instrumentalness`, and `liveness`, are all within their expected ranges.

We also found 157 rows with `tempo` equal to 0 and 68 rows with positive `loudness`. We keep these rows for now because they are a small part of the dataset and may just be unusual Spotify estimates rather than clear errors. We will be careful when using these variables later.

### Dataset #2: Most Streamed Spotify Songs 2024 + Dataset Merging

This dataset contains 4,600 highly streamed songs from 2024, along with cross-platform performance metrics from Spotify, YouTube, TikTok, Apple Music, Deezer, Amazon, Pandora, SiriusXM, Soundcloud, and Shazam. In our project, this dataset is not the main audio-feature dataset. Instead, we use it to identify recent hit songs and then compare those matched hit songs with the broader Spotify Tracks sample.

**Key variables and their meanings:**

- `Track`, `Artist`, `Album Name`, `Release Date`: Track identifiers. `Track` and `Artist` are especially important because we use them to match songs to the Spotify Tracks Dataset.
- `ISRC`: International Standard Recording Code. This is a unique global identifier for recordings, but we do not use it as the main merge key because the Spotify Tracks Dataset does not include the same ISRC column.
- `All Time Rank`: Ranking variable provided by the dataset for the listed songs.
- `Track Score`: A dataset-provided score related to cross-platform performance. Since the exact formula is not fully explained, we treat it as a descriptive variable instead of our main outcome variable.
- `Spotify Streams`: Total Spotify stream count. In the raw file, this column is stored as a comma-formatted string, so it needs to be converted into numeric form during cleaning.
- `Spotify Popularity`: Spotify popularity score from 0 to 100. This is useful for describing the 2024 hit songs, but it has missing values.
- `YouTube Views`, `TikTok Views`: View or play counts on those platforms. These variables help describe cross-platform popularity, but they also have missing values.
- `Explicit Track`: A 0 or 1 flag indicating whether the track contains explicit content.

**Missing data:** This dataset has many missing values in platform-specific columns. We do not drop every row with missing values because many of those columns are not needed for our main analysis or for matching songs. Missing values are kept as `NaN` instead of being filled with 0, because missing data means the value was not recorded, not that the true value is zero. For example, a missing value in `TikTok Views` does not necessarily mean the song had zero TikTok views.

We drop `TIDAL Popularity` because it is completely missing. We also drop rows that are missing `Track` or `Artist`, because those rows cannot be matched reliably to the Spotify Tracks Dataset. Other platform metrics, such as `Soundcloud Streams`, `Pandora Streams`, and `SiriusXM Spins`, are kept only as optional descriptive variables and are not used as main predictors. Numeric columns such as `Spotify Streams`, `Spotify Popularity`, `YouTube Views`, and `TikTok Views` are converted from comma-formatted strings into numeric values.

**Dataset concerns:** This dataset is biased toward globally popular and mainstream songs from 2024. It does not represent all Spotify songs, and it also does not represent less-known songs, older songs, or independent artists very well. Another important limitation is that this dataset does not include the full Spotify audio features used in our main regression analysis. Because of that, it cannot replace the Spotify Tracks Dataset. Its role is only to help us identify recent hit songs for a supplementary comparison.

**Merging strategy:** We merge this dataset with the cleaned Spotify Tracks Dataset using a normalized inner join on `track_name` and `artist`, where both fields are lowercased and stripped of extra spaces. Songs that cannot be matched are excluded from the merged subset. Since the 2024 dataset does not contain the full audio-feature profile, matched songs get their audio features from the cleaned Spotify Tracks Dataset.

The current merge produces 1,311 matched rows. However, this does not mean 1,311 unique hit songs, because duplicate entries, repeated tracks, different versions, or slightly different track records can create multiple matched rows. We use this merged dataset only as a supplementary comparison. The main analysis still focuses on the relationship between Spotify audio features and popularity score in the cleaned Spotify Tracks Dataset. The merged dataset is saved as `data/02-processed/spotify_merged_clean.csv`.

```python
import pandas as pd
import numpy as np
import os

# --- Load raw 2024 hits data ---
hits_raw = pd.read_csv('data/00-raw/spotify_2024.csv', encoding='latin1')
print('Raw shape:', hits_raw.shape)

print('\nMissing values per column:')
missing = hits_raw.isnull().sum()
print(missing[missing > 0].sort_values(ascending=False))

# --- Select relevant columns ---
# We do not keep TIDAL Popularity because it is 100% missing.
cols_keep = [
    'Track', 'Artist', 'Album Name', 'Release Date', 'ISRC',
    'All Time Rank', 'Track Score',
    'Spotify Streams', 'Spotify Popularity',
    'YouTube Views', 'TikTok Views', 'Explicit Track'
]

hits = hits_raw[cols_keep].copy()

# --- Drop rows that cannot be matched ---
# Track and Artist are needed for matching with the Spotify Tracks Dataset.
hits = hits.dropna(subset=['Track', 'Artist']).reset_index(drop=True)

# --- Convert comma-formatted numeric columns to numeric ---
# Missing values stay as NaN. We do not fill them with 0.
numeric_cols = [
    'All Time Rank', 'Track Score',
    'Spotify Streams', 'Spotify Popularity',
    'YouTube Views', 'TikTok Views',
    'Explicit Track'
]

for col in numeric_cols:
    hits[col] = (
        hits[col]
        .astype(str)
        .str.replace(',', '', regex=False)
    )
    hits[col] = pd.to_numeric(hits[col], errors='coerce')

print('\n2024 hits shape after cleaning:', hits.shape)

print('\nMissing values after cleaning:')
missing_clean = hits.isnull().sum()
print(missing_clean[missing_clean > 0].sort_values(ascending=False))

# --- Save cleaned 2024 dataset ---
os.makedirs('data/02-processed', exist_ok=True)
hits.to_csv('data/02-processed/spotify_2024_clean.csv', index=False)
print('\nSaved: data/02-processed/spotify_2024_clean.csv')

# --- Load cleaned tracks dataset ---
tracks = pd.read_csv('data/02-processed/spotify_tracks_clean.csv')

# --- Normalize keys for matching ---
# Simple normalization: lowercase and remove extra spaces.
hits['_track_key'] = hits['Track'].str.lower().str.strip()
hits['_artist_key'] = hits['Artist'].str.lower().str.strip()

tracks['_track_key'] = tracks['track_name'].str.lower().str.strip()
tracks['_artist_key'] = tracks['artists'].str.lower().str.strip()

# --- Inner join on normalized track + artist ---
merged = pd.merge(
    hits,
    tracks,
    on=['_track_key', '_artist_key'],
    how='inner'
)

# Drop temporary matching keys
merged = merged.drop(columns=['_track_key', '_artist_key'])

print('\nMerged shape:', merged.shape)
print('Unique matched Track + Artist pairs:', merged[['Track', 'Artist']].drop_duplicates().shape[0])
print('Unique matched track_id values:', merged['track_id'].nunique())

print('\nMissing values in merged dataset:')
missing_m = merged.isnull().sum()
print(missing_m[missing_m > 0].sort_values(ascending=False))

# --- Save merged dataset ---
merged.to_csv('data/02-processed/spotify_merged_clean.csv', index=False)
print('\nSaved: data/02-processed/spotify_merged_clean.csv')
```
